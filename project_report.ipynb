# Final Project Solution Report

The intent of this project was to provide valueable churn insight to Interconnect telecom operator using data from users of their services. The data was analyzed in terms of what factors contribute to Interconnect users discontinuing use of services. By the data, a binary classification machine learning model was trained so that Interconnect can use it to predict which of their current users are likely to leave at any given time. The goal of the project is to maximize a model's AUC-ROC score, a metric ranging from 0 to 1 that measures how well a classification model distinguishes between classes.

All steps recorded in this project's work plan were followed. The main challenge when completing the project was tuning and testing several models that needed differently encoded data in a timely manner. This was managed by using pipelines that encoded, scaled, and defined particular models as well as using GridSearchCV on relatively small parameter grids to start. If that was too time and memory intensive, RandomizedSearchCV would have been used on larger parameter grids. 

LinearRegression, RandomForest, XGBoost, and LightGBM models were trained and evaluated on customer data based on accuracy and AUC-ROC score. The optimal model for predicting whether or not an Interconnect customer will churn was the LightGBM model with the below parameters.

{'classifier__learning_rate': 0.1, 'classifier__max_depth': -1, 'classifier__n_estimators': 200, 'classifier__num_leaves': 50}

This LightGBM model returned an AUC-ROC score of 0.89 along with an accuracy score of 0.84, the highest scores of all models trained and evaluated and sufficient for the needs of Interconnect's churn prediction needs.
